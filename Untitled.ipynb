```python
import pandas as pd
import lightgbm as lgb
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

Import data

```python
housing_df = pd.read_csv('./ml_house_data_set.csv')
housing_df.head()
```

```python
del housing_df['house_number']
del housing_df['street_name']
del housing_df['unit_number']
del housing_df['zip_code']
```

```python
garage_df=pd.get_dummies(housing_df['garage_type'])

housing_df = housing_df.assign(attached=garage_df['attached'].values)
housing_df = housing_df.assign(attached=garage_df['detached'].values)
housing_df = housing_df.assign(attached=garage_df['none'].values)

del housing_df['garage_type']
del garage_df
```

```python
city_df=pd.get_dummies(housing_df['city'])
housing_df = housing_df.join(city_df)
del housing_df['city']
del city_df
```

```python
housing_df.head()
```

```python
y = housing_df['sale_price']
y
```

```python
X = housing_df.drop('sale_price', 1)
X.head()
```

```python
housing_df.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=4)
```

```python
gbm = lgb.LGBMRegressor(objective='regression',
                        num_leaves=80,
                        learning_rate=1,
                        n_estimators=40)
```

```python
gbm.fit(X_train, y_train,
        eval_set=[(X_test, y_test)],
        eval_metric='l2',
        early_stopping_rounds=10)
```

```python
y_pred = gbm.predict(X_test, num_iteration=gbm.best_iteration)
```

```python
print('The rmse of prediction is:', mean_squared_error(y_test, y_pred) ** 0.5)
```

```python
print('Feature importances:', list(gbm.feature_importances_))
```

```python
estimator = lgb.LGBMRegressor(num_leaves=80)

param_grid = {
    'learning_rate': [0.01, 0.1, 1],
    'n_estimators': [20, 40]
}
```

```python
gbm = GridSearchCV(estimator, param_grid)
```

```python
gbm.fit(X_train, y_train)
```

```python
print('Best parameters found by grid search are:', gbm.best_params_)
```
